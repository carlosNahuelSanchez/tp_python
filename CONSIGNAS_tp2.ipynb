### 1. **Bucles**

   - **Actividad 1.1**: Escribir un programa que muestre el print de todo lo que el usuario introduzca hasta que el usuario escriba “salir” que terminará.
   
   - **Actividad 1.2**: Escribir un programa que pida al usuario un número entero y muestre por pantalla un triángulo rectángulo como el de más abajo con la cantidad de escalones indicados previamente (los numeros dentro del triangulo deben ser enteros aleatorios del 0 al 9).
* 1
* 3 1
* 5 3 1
* 7 5 3 1
* 9 7 5 3 1

```python

def printUserInputs():
    print("Escribe lo que deseas para verlo en terminal :D. Si quieres dejar de escribir solo escribe QUIT")
    while True: # El bucle se va a repetir indefinidamente hasta que se cumpla una condición
        inputUser = input("Ingrese texto: ") # Se solicita al usuario que ingrese un input
        if inputUser == "QUIT": # Condición que termina el bucle
            break 
        print(f"Ingresaste: {inputUser}"); # Se imprime el input del usuario

printUserInputs()
```

```python
import random # Librería para generar numeros aleatorios

def triangleOfNumbers(): 
        # Se le solicita al usuario el numero de filas de su triangulo
        n = int(input("Ingresar un número entero para definir las filas tu triangulo: ")) 
        for i in range(1, n + 1): # Bucle externo: Controla la cantidad de filas del triangulo
                for j in range(i): # Bucle interno: Controla la cantidad de numeros de cada fila
                        randomNumber = random.randint(0, 9) # Generar número aleatorio del 0 al 9
                        print(randomNumber, end=" ") # Evita el salto de linea en cada bucle
        print()
        
triangleOfNumbers()
```

### 2. **Listas y Tuplas**

```python
def vocalsAndSpacesDetector():
    
    phraseUser = input("Ingrese una oración") # Se le solicita al usuario una oración
    
    phraseUserUpper = phraseUser.upper() # Se convierte toda la oración a mayuscula para facilitar la busqueda de vocales
    
    # Se cuentan las apariciones de las vocales y espacios en la oración con el metódo count()
    numbers_A = phraseUserUpper.count("A")
    numbers_E = phraseUserUpper.count("E")
    numbers_I = phraseUserUpper.count("I")
    numbers_O = phraseUserUpper.count("O")
    numbers_U = phraseUserUpper.count("U")
    numbers_Spaces = phraseUserUpper.count(" ")

    # Se imprime la respuesta
    print(f"La letra A aparece: {numbers_A} ")
    print(f"La letra E aparece: {numbers_E} ")
    print(f"La letra I aparece: {numbers_I} ")
    print(f"La letra O aparece: {numbers_O} ")
    print(f"La letra U aparece: {numbers_U} ")
    print(f"La frase tiene: {numbers_Spaces} Espacios ")

        
vocalsAndSpacesDetector()
```

```python
import random # Librería para generar numeros aleatorios

def matrixGenerator():
    matrixRows = int(input("Ingrese la cantidad de filas de su matriz"))
    matrixColumns = int(input("Ingrese la cantidad de filas de su matriz"))
    
    matriz = [[random.randint(1, 100) for _ in range(matrixColumns)] for _ in range(matrixRows)]

    print("Matriz Normal:")
    for row in matriz: print(row) # Se imprime la matriz generada
    
    # Se transpone la matriz usando zip y se convierte en una lista de listas
    matriz_transpuesta = [list(fila) for fila in zip(*matriz)]

    # Se tranforma la matriz transpuesta en una tupla de tuplas
    tupla_transpuesta = tuple(matriz_transpuesta)
    
    # Se imprime la matriz transpuesta  
    print("\nMatriz transpuesta (en tupla):")
    for fila in tupla_transpuesta:
        print(fila)
    
    
matrixGenerator()
```

```python
def menorAMayorPrecio():
    
    precios = [50, 75, 46, 22, 80, 65, 8, 100, 12, 90] # Lista de precios
    
    precioMayor = max(precios) # Se obtiene el precio mayor de la lista
    precioMenor = min(precios) # Se obtiene el precio menor de la lista
    
    print(f"El precio mayor es: {precioMayor}") # Se imprime el precio mayor
    print(f"El precio menor es: {precioMenor}") # Se imprime el precio menor
    

menorAMayorPrecio()
```

### 3. **Diccionarios**

   - **Actividad 1**: Escribir un programa que cree un diccionario simulando una cesta de la compra. El programa debe preguntar el artículo y su precio y añadir el par al diccionario, hasta que el usuario decida terminar. Después se debe mostrar por pantalla la lista de la compra y el coste total, con el siguiente formato.

   | arts  |precio |
   |-------|-------|
   |art 1  |1      |
   |art 2  |2      |
   |art 3  |3      |
   |total  |6      |

   - **Actividad 2**: Escribir un programa que cree un diccionario de traducción español-inglés. El usuario introducirá las palabras en español e inglés separadas por dos puntos, y cada par <palabra> : <traducción> separados por comas (es decir, solo se debe usar **UNA VEZ** la función input para la carga de palabras). El programa debe crear un diccionario con las palabras y sus traducciones. Después pedirá una frase en español y utilizará el diccionario para traducirla palabra a palabra. Si una palabra no está en el diccionario debe dejarla sin traducir.

   - **Actividad 3**: Escribir un programa que reciba una cadena de caracteres y devuelva un diccionario con cada palabra que contiene y su frecuencia. Escribir otra función que reciba el diccionario generado con la función anterior y devuelva una tupla con la palabra más repetida y su frecuencia.

```python

def cestaCompras():
    # Creamos el diccionario para la cesta
    cesta = {}

    # Iniciamos un bucle para que el usuario agregue productos
    while True:
        articulo = input("Ingrese el nombre del artículo (o escriba 'fin' para terminar): ")
        
        # Si el usuario quiere terminar
        if articulo.lower() == 'fin':
            break

        # Pedimos el precio
        try:
            precio = float(input(f"Ingrese el precio de '{articulo}': "))
        except ValueError:
            print("Precio inválido. Intente nuevamente.")
            continue
        
        # Se agrega el artículo al diccionario
        cesta[articulo] = precio

    # Se calcula el total
    total = sum(cesta.values())

    # Mostramos los resultados en forma de tabla
    print("\n{:10s} | {:>6s}".format("arts", "precio"))
    print("-" * 20)

    for art, precio in cesta.items():
        print("{:10s} | {:6.2f}".format(art, precio))

    # Imprimimos la fila final con el total
    print("{:10s} | {:6.2f}".format("total", total))
    
cestaCompras()
```

```python

def diccionarioIngles():
    # Pedimos al usuario que cargue los pares palabra:traducción
    entrada = input("Ingrese los pares <palabra>:<traducción> separados por comas: ")

    # Creamos el diccionario vacío
    diccionario = {}

    # Dividimos la entrada en cada par por la coma
    pares = entrada.split(",")

    # Recorremos cada par para separarlos por los dos puntos
    for par in pares:
        # Quitamos espacios en blanco al principio o final
        par = par.strip()

        # Separamos por los dos puntos (:) y validamos
        if ":" in par:
            espanol, ingles = par.split(":")
            diccionario[espanol.strip()] = ingles.strip()
        else:
            print(f"Formato inválido en: {par}")

    # Mostramos el diccionario para verificar
    print("\nDiccionario cargado:")
    print(diccionario)

    # Pedimos una frase en español
    frase = input("\nIngrese una frase en español para traducir: ")

    # Separamos la frase en palabras
    palabras = frase.split()

    # Traducimos palabra por palabra
    traduccion = []
    for palabra in palabras:
        # Si existe en el diccionario, usamos su traducción
        if palabra in diccionario:
            traduccion.append(diccionario[palabra])
        else:
            traduccion.append(palabra)  # Dejamos la original si no existe

    # Unimos y mostramos la frase traducida
    frase_traducida = " ".join(traduccion)
    print("\nFrase traducida:")
    print(frase_traducida)

diccionarioIngles()
```

```python
def contar_palabras():
    
    texto = input("Ingresá una cadena de texto: ")  # Se le solicita al usuario que ingrese una cadena de texto
    
    palabras = texto.split()  # Se separa el texto en una lista de palabras usando espacios como separador
    
    frecuencias = {}  # Se crea un diccionario vacío para guardar la frecuencia de cada palabra

    for palabra in palabras:
        palabra = palabra.lower()  # Se convierte cada palabra a minúsculas
        if palabra in frecuencias:
            frecuencias[palabra] += 1  # Si la palabra ya existe en el diccionario, se incrementa su contador
        else:
            frecuencias[palabra] = 1  # Si es la primera vez que aparece, se agrega al diccionario con valor 1

    print("Diccionario de frecuencias:", frecuencias)  # Se imprime el diccionario con las palabras y sus frecuencias
    return frecuencias  # Se retorna el diccionario para usarlo en la siguiente función



def palabra_mas_repetida():
    
    diccionario = contar_palabras()  # Se llama a la función contar_palabras y se guarda el diccionario resultante
    
    max_palabra = None  # Variable para guardar la palabra más repetida
    max_frecuencia = 0  # Variable para guardar la mayor frecuencia encontrada

    for palabra, frecuencia in diccionario.items():
        if frecuencia > max_frecuencia:
            max_palabra = palabra  # Se actualiza la palabra más repetida
            max_frecuencia = frecuencia  # Se actualiza la frecuencia máxima

    print("Palabra más repetida y su frecuencia:", (max_palabra, max_frecuencia))  # Se imprime la palabra más repetida y cuántas veces apareció
    

palabra_mas_repetida() 
```

### 4. **Funciones**

- **Actividad 1**: Escribir una función que calcule el total de una factura tras aplicarle el IVA. La función debe recibir la cantidad sin IVA y el porcentaje de IVA a aplicar, y devolver el total de la factura. Si se invoca la función sin pasarle el porcentaje de IVA, deberá aplicar un 21%.

- **Actividad 2**: Escribir una función que reciba un número entero positivo y devuelva su factorial.


```python
def calcular_total_con_iva():
    
    precio_sin_iva = float(input("Ingrese el precio sin IVA: "))
    
    # Definimos el porcentaje de IVA
    porcentaje_iva = (input("Ingrese el porcentaje de IVA: "))
    
    # Verificamos si se ingresó un valor, si no, usamos 21
    if porcentaje_iva == "":
        porcentaje_iva = 21
    else:
        porcentaje_iva = float(porcentaje_iva)
        
    # Calculamos el monto del IVA
    iva = precio_sin_iva * (porcentaje_iva / 100)

    # Sumamos el IVA al precio original
    total = precio_sin_iva + iva

    print(f"El precio total con IVA es: {total}")
    
calcular_total_con_iva()
```

```python
def factorial():
    
    n = int(input("Ingrese un número entero positivo para calcular su factorial: "))
    
    # Validamos que el número sea entero y positivo
    if n < 0:
        return "El número debe ser un entero positivo"

    resultado = 1  # El factorial de 0 es 1, por eso iniciamos con 1

    # Multiplicamos desde 1 hasta n
    for i in range(1, n + 1):
        resultado *= i

    print(resultado)

factorial()
```
